# Visualisation and Insights from Rural Health Statistics 2016

As on 31st March, 2015, there were 153655 Sub Centres, 25308 Primary
Health Centres (PHCs) and 5396 Community Health Centres (CHCs)
functioning in the country. While the Sub Centres, PHCs and CHCs have
increased in number in 2014-15, the current numbers are not sufficient to
meet their population norm.

```python
import pandas as pd

import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

# A Look on the Statistics of Building of Health Centers

```python
train = pd.read_csv("demo.csv") # the iris dataset is now a Pandas DataFrame


train
```

# Functioning CHCs as of 2016, March 31st

```python
train.plot(kind="barh", x="State/UT", y="Total Number of CHCs functioning - (As on 31st March, 2016)",colors='r')
plt.xlabel('Functioning CHCs')
plt.show()
```

# Functioning CHCs in government Buildings

```python
train.plot(kind="barh", x="State/UT", y="CHCs functioning in Govt. Builidings - (As on 31st March, 2016)",colors='g')
plt.xlabel('Functioning CHCs in Government Buildings')
plt.show()
```

# Functioning CHCs in Rent Free Panchayat Buildings

```python
train.plot(kind="barh", x="State/UT", y="CHCs functioning in Rent Free Panchayat/ Vol. Society Buildings - (As on 31st March, 2016)")
plt.xlabel('Functioning CHCs in Rent Free Panchayat')
plt.show()
```

# Underconstruction Buildings Statewise

```python
train.plot(kind="barh", x="State/UT", y="Buildings Under Construction - (As on 31st March, 2016)")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1))
plt.xlabel('CHCs Buildings Under Construction')
plt.show()
```

# Buildings Required to be constructed

```python
train.plot(kind="barh", x="State/UT", y="Buildings required to be constructed - (As on 31st March, 2016)")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1))
plt.xlabel('CHCs Buildings Required to be constructed')
plt.show()
```

```python

e = {'Region' : ['Panchayat','government','Underconstruction'],'Numbers' :[train['CHCs functioning in Rent Free Panchayat/ Vol. Society Buildings - (As on 31st March, 2016)'].sum(),
                                                                train['CHCs functioning in Govt. Builidings - (As on 31st March, 2016)'].sum(),
                                                               train['Buildings Under Construction - (As on 31st March, 2016)'].sum()]
     }
```

# Overall Numers of buildings in different catagories

```python
df=pd.DataFrame(e)
df
```

```python
df.plot(kind="bar", x="Region", y="Numbers")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1))
plt.xlabel('Regions')
plt.show()
```

```python
labels = 'Panchayat', 'government', 'Underconstruction'
sizes = df['Numbers']
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
